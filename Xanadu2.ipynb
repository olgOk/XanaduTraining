```
import strawberryfields as sf
from strawberryfields.apps import  data, plot, sample
import numpy as np
```

```
from strawberryfields.apps import data, sample

samples = data.Planted()

A = samples.adj
n_mean = samples.n_mean
more_samples = sample.sample(A, n_mean, n_samples=5)
print(more_samples)
```

```
import networkx as nx

samples = data.Planted()
g = nx.Graph(samples.adj)
samples = sample.postselect(samples, 16, 30)
samples = sample.to_subgraphs(samples, g)
plot.graph(g, samples[7])
```

```
from strawberryfields.apps import clique

c = clique.shrink(samples[2],g)
plot.graph(g,c)
```

```
c = clique.grow(c,g)
plot.graph(g,c)
```

```
c = clique.swap(c,g)
plot.graph(g, c)
```

```
shrunk_cliques = [clique.shrink(s, g) for s in samples]
search_cliques = [clique.search(s, g, 10) for s in shrunk_cliques]
cliques_plus_lengths = [(len(s), s) for s in shrunk_cliques]
cliques_plus_lengths = sorted(cliques_plus_lengths, reverse=True)
print(cliques_plus_lengths[0])
```

```
samples = data.TaceAs()

g = nx.Graph(samples.adj)
samples = sample.postselect(samples, 8,8)
samples = sample.to_subgraphs(samples, g)

cliques = [clique.shrink(s, g) for s in samples]
cliques = [clique.search(s, g, 10) for s in cliques]

print(max(len(c) for c in cliques))
```

```
from strawberryfields.apps import data, sample, vibronic

f = data.Formic()

params_0 = vibronic.gbs_params(f.w, f.wp, f.Ud, f.delta, T=0)
params_300 = vibronic.gbs_params(f.w, f.wp, f.Ud, f.delta, T=300)

samples_0 = sample.vibronic(*params_0, n_samples=10)
samples_300 = sample.vibronic(*params_300, n_samples=10)

energies_0 = vibronic.energies(samples_0, f.w, f.wp)
energies_300 = vibronic.energies(samples_300, f.w, f.wp)
```
